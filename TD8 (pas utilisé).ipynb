```python
import pandas as pd

df = pd.read_csv("C:/Users/Thibaud/Desktop/Programmation de spécialité Python/discours_US.csv", sep = "\t")
print(df)
speaker = df["speaker"].value_counts()
print(speaker)
from Corpus import Corpus
from TD4_Classes import Document
import datetime
Discours_corpus = Corpus("Discours")
collection = []
print(df.columns)
for i, entry in enumerate(df['text']):
    phrase = entry.split('.')
    author = df.loc[i, "speaker"]
    date = df.loc[i, "date"]
    date_obj = datetime.datetime.strptime(date, "%B %d, %Y")
    formatted_date = date_obj.strftime("%Y/%m/%d")
    titre = df.loc[i, "descr"]
    url = df.loc[i, "link"]
    type = 'discours'
    for y in phrase:
        doc = Document(titre, author, formatted_date, url, y, type)
        collection.append(doc)
```

```python
from tqdm import tqdm

for doc in collection:
    Discours_corpus.add(doc)

# Recherche d'un mot spécifique dans le corpus
mot_cle = "Thank"
resultats_search = Discours_corpus.search(mot_cle)
resultats_concorde = Discours_corpus.concorde(mot_cle, 50)

# Affichage des résultats
print(f"Occurrences du mot '{mot_cle}' trouvées dans le corpus :")
for doc_id, mot_trouve in tqdm(resultats_search):
    print(f"Document ID: {doc_id}, Mot trouvé: {mot_trouve}")

print(f"Occurrences du mot '{mot_cle}' trouvées dans le corpus :")
print(resultats_concorde)


print(Discours_corpus.vocabulaire())
print(Discours_corpus.calculer_freq())
        
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from Corpus import Corpus
from Search_Engine import SearchEngine
import pickle

# Étape 3.1 : Création des widgets principaux
label = widgets.Label("Interface de Recherche")
text_input = widgets.Text(description="Mots-clés:", placeholder="Entrez les mots-clés...")
int_slider = widgets.IntSlider(value=5, min=1, max=20, step=1, description="Résultats:")

# Étape 3.3 : Création de l'objet Output pour afficher les résultats
output = widgets.Output()

# Étape 3.4 : Création du bouton avec une action personnalisée
search_button = widgets.Button(description="Lancer la recherche")


# Charger le corpus sauvegardé au format pickle meilleur méthode pour importer/créer la classe que csv
corpus_pickle = Corpus.from_pickle("corpus.pkl")
#plus rapide avec corpus_pickle car plus petit
moteur = SearchEngine(Discours_corpus)

# Étape 3.5 : Définition de la fonction qui gère le clic sur le bouton
def on_button_click(b):
    with output:
        clear_output()
        query = text_input.value
        num_results = int_slider.value
        
        if not query:
            print("Veuillez entrer des mots-clés pour la recherche.")
        else:
            results = moteur.search(query, num_results)
            print(results)

# Lier la fonction au bouton
search_button.on_click(on_button_click)

# Étape 3.2 : Organisation de l'interface avec VBox et HBox
ui = widgets.VBox([
    label,
    text_input,
    int_slider,
    search_button,
    output
])

# Afficher l'interface
display(ui)
```
